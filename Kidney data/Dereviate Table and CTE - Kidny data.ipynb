# Queries with derived tables (I)


A derived table is a query which is used in the place of a table. Derived tables are a great solution if you 
<br>
want to create intermediate calculations that need to be used in a larger query.
<br>
Here we want to calculate the maximum value of the blood glucose level for each record by age.


### what is the maximum value of the blood glucose level for each record by age.

```python
%%sql
SELECT a.RecordId, a.Age, a.BloodGlucoseRandom,   
       b.*
FROM Kidney a
join (SELECT Age, MAX(BloodGlucoseRandom) AS MaxGlucose FROM Kidney GROUP BY Age) b
on a.Age=b.Age 
```

### return all patient records with the highest BloodPressure at their Age level

```python
%%sql
SELECT *
FROM Kidney a
-- Create derived table: select age, max blood pressure from kidney grouped by age
JOIN (select age, MAX(BloodPressure) AS MaxBloodPressure from kidney group by age) b
-- JOIN on BloodPressure equal to MaxBloodPressure
ON a.BloodPressure=b.MaxBloodPressure
-- Join on Age
AND a.age=b.age
```

# Common table expression

### return all the ages with the maximum BloodGlucoseRandom in the table by using CTE 

1- create CTE named BloodGlucoseRandom
<br>
2-make a join of the cte to the kidney table 

```python
%%sql

with BloodGlucoseRandom (MaxGlucose) 
AS (SELECT MAX(BloodGlucoseRandom) AS MaxGlucose FROM Kidney)

SELECT a.Age, b.MaxGlucose
FROM Kidney a
-- Join the CTE on blood glucose equal to max blood glucose
JOIN BloodGlucoseRandom b
on a.BloodGlucoseRandom=b.MaxGlucose
```

### return all the information regarding the patient(s) with the maximum BloodPressure

```python
%%sql
-- Create the CTE
WITH BloodPressure(MaxBloodPressure) 
AS (SELECT MAX(BloodPressure) AS MaxBloodPressure FROM Kidney)

SELECT *
FROM Kidney a
join BloodPressure b
on a.BloodPressure=b.MaxBloodPressure
```
